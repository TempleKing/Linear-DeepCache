```python
"""
从大 CSV 中切出一小部分用于实验。
上传一次大文件，之后只用小文件跑实验。
"""

import csv
import random
import os


def split_csv(
    input_path: str,
    output_path: str = "prompts_small.csv",
    prompt_column: str = "prompt",
    n: int = 50,
    shuffle: bool = True,
    seed: int = 30,
):
    """
    从大 CSV 切出 n 条保存为小 CSV。

    Args:
        input_path:    原始大 CSV 路径
        output_path:   输出小 CSV 路径
        prompt_column: prompt 所在列名
        n:             切多少条
        shuffle:       是否随机抽样（False = 取前 n 条）
        seed:          随机种子
    """
    # 读取
    with open(input_path, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        if prompt_column not in reader.fieldnames:
            raise ValueError(
                f"列 '{prompt_column}' 不存在，"
                f"可用列: {reader.fieldnames}"
            )
        rows = [row for row in reader if row[prompt_column].strip()]

    print(f"原始: {len(rows)} 条")

    # 抽样
    if shuffle:
        random.seed(seed)
        rows = random.sample(rows, min(n, len(rows)))
    else:
        rows = rows[:n]

    print(f"切出: {len(rows)} 条")

    # 写入
    with open(output_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=[prompt_column])
        writer.writeheader()
        for row in rows:
            writer.writerow({prompt_column: row[prompt_column].strip()})

    print(f"保存: {output_path}")
    print(f"大小: {os.path.getsize(output_path)} bytes")

    # 预览
    print(f"\n前 5 条预览:")
    for i, row in enumerate(rows[:5]):
        p = row[prompt_column].strip()
        print(f"  {i+1}. {p[:80]}{'...' if len(p) > 80 else ''}")


if __name__ == "__main__":
    split_csv(
        input_path="diffusion_prompts.csv",
        output_path="prompts_LITTLE4.csv",
        prompt_column="prompt",
        n=50,
    )
```
